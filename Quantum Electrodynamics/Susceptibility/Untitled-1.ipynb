```python
import numpy as np
import matplotlib.pyplot as plt

# Constants (arbitrary values, adjust as needed)
N = 1e19  # Number density (atoms per m^3)
mu_13 = 1e-29  # Dipole moment (Cm)
Omega_13 = 1e6  # Rabi frequency (Hz)
Omega_23 = 1e6  # Rabi frequency (Hz)
Gamma_2 = 1e6  # Decay rate for level 2 (Hz)
Gamma_3 = 1e6  # Decay rate for level 3 (Hz)
epsilon_0 = 8.854e-12  # Vacuum permittivity (F/m)
hbar = 1.055e-34  # Reduced Planck constant (Js)

# Frequency detuning values (omega_p)
omega_c = 2e15  # Control field frequency (Hz), adjust as needed
omega_p = np.linspace(omega_c - 10e6, omega_c + 10e6, 1000)  # Probe field frequencies around control field

# Detuning Delta_2 = omega_p - omega_c
Delta_p = omega_p - omega_c

# Define gamma_2 and gamma_3
gamma_2 = Gamma_2 - 1j * Delta_p  # Decay rate for state 2 (complex)
gamma_3 = Gamma_3 - 1j * 0  # Decay rate for state 3 (complex, assuming no detuning for level 3)

# Susceptibility formula
real_chi = - (N * mu_13**2 / (hbar * epsilon_0 * (gamma_2 * gamma_3 + Omega_23**2))) * (Omega_13 * Omega_23 - gamma_2.real)
imag_chi = - (N * mu_13**2 / (hbar * epsilon_0 * (gamma_2 * gamma_3 + Omega_23**2))) * (Omega_13 * Omega_23 - gamma_2.imag)

# Plotting the real and imaginary parts of susceptibility
plt.figure(figsize=(12, 8))

# Plot real part of susceptibility vs omega_p
plt.subplot(2, 1, 1)
plt.plot(omega_p, real_chi, label="Real Part", color='blue')
plt.xlabel(r'$\omega_p$ (Hz)')
plt.ylabel(r'Re[$\chi(\omega_p)$]')
plt.title('Real Part of Susceptibility $\chi(\omega_p)$ vs $\omega_p$')
plt.grid(True)

# Plot imaginary part of susceptibility vs omega_p
plt.subplot(2, 1, 2)
plt.plot(omega_p, imag_chi, label="Imaginary Part", color='red')
plt.xlabel(r'$\omega_p$ (Hz)')
plt.ylabel(r'Im[$\chi(\omega_p)$]')
plt.title('Imaginary Part of Susceptibility $\chi(\omega_p)$ vs $\omega_p$')
plt.grid(True)

plt.tight_layout()
plt.show()
```
